```python
import csv 
from wordcloud import WordCloud 
import pandas as pd
```

```python
data=pd.read_csv("SPAM text message 20170820 - Data.csv",encoding='utf-8')
```

```python
spam=data.loc[data["Category"]=="spam",["Message"]]
```

```python
listS=[]
for i in spam.Message:
    listS.append(i)
    
```

```python
print(type(listS[0]))
```

```python
wordcloud = WordCloud().generate(str(listS))
```

```python
import matplotlib.pyplot as plt 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
```

```python
ham=data.loc[data["Category"]=="ham",["Message"]]
listH=[]
for i in ham.Message:
    listH.append(i)
    
wordcloudH = WordCloud().generate(str(listH))
plt.imshow(wordcloudH, interpolation='bilinear') 
plt.axis("off") 
```
