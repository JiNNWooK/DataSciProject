```python
listMessage=listMessage.Message
```

```python
newListMsg=[]
for i in listMessage:
    sublist=[]
    i=re.sub('[^a-zA-Z]', ' ',i)
    i=i.lower().split()
    stops = set(stopwords.words('english'))
    no_stops = [word for word in i if not word in stops]
    stemmer = nltk.stem.SnowballStemmer('english')
    stemmer_words = [stemmer.stem(word) for word in no_stops]
    sublist.append(stemmer_words)
    newListMsg.append(sublist)
```

```python
tokenlist=[]
for i in listMessage:
    tokens=nltk.word_tokenize(i)
    tokenlist.append(tokens)
```

```python
poslist=[]
for i in tokenlist:
        tag=nltk.pos_tag(i)
        poslist.append(tag)
```
